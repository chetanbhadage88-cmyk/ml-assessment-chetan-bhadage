```python
#Task 1
# Import libraries
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Load dataset
df = pd.read_csv('q2_customers.csv')

# Display first few rows
print(df.head())

# Initialize scaler
scaler = StandardScaler()

# Scale all features
scaled_data = scaler.fit_transform(df)

# Convert back to DataFrame for readability
df_scaled = pd.DataFrame(scaled_data, columns=df.columns)

# Display scaled data
print(df_scaled.head())
```

```python
#Task 1 Justification:
# Scaling the data is crucial for many machine learning algorithms, especially those that rely on distance metrics (like KNN, SVM, etc.) or gradient-based optimization (like logistic regression, neural networks). By standardizing the features to have a mean of 0 and a standard deviation of 1, we ensure that all features contribute equally to the model's performance and prevent any single feature from dominating due to its scale. This can lead to improved convergence during training and better overall model performance. 
```

```python
#Task 2
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# WCSS list
wcss = []

# Try K from 1 to 10
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(df_scaled)
    wcss.append(kmeans.inertia_)

# Plot the Elbow graph
plt.figure(figsize=(8,5))
plt.plot(range(1, 11), wcss, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.title('Elbow Method for Optimal K')
plt.show()
```

```python
#Task 2 justification:
# The Elbow method is a widely used technique to determine the optimal number of clusters in K-means clustering. By plotting the WCSS against the number of clusters, we can visually identify the point where adding more clusters does not significantly reduce the WCSS, which is often referred to as the "elbow". This point indicates a good balance between model complexity and performance, helping us choose an appropriate K for our clustering task.        
```

```python
#Task 3
from sklearn.cluster import KMeans

# Choose optimal K (replace 3 with your elbow result if different)
optimal_k = 3

# Fit K-Means
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
clusters = kmeans.fit_predict(df_scaled)

# Add cluster column to original dataframe
df['cluster'] = clusters

# View clustered data
print(df.head())

# Centroids (convert back to DataFrame for readability)
centroids = pd.DataFrame(
    kmeans.cluster_centers_,
    columns=df.columns[:-1]  # exclude cluster column
)

print("\nCluster Centroids (Scaled Values):")
print(centroids)
```

```python
#Task 3 justification:
# Fitting K-Means with the optimal number of clusters allows us to segment the customers into distinct groups based on their features. By adding the cluster labels to the original dataframe, we can analyze the characteristics of each cluster and identify patterns or insights about customer behavior. The centroids provide a summary of the average feature values for each cluster, which can help in understanding the defining characteristics of each group and inform targeted marketing strategies or business decisions.
#cluster 0 means that the customers in this cluster have lower values for all features compared to the other clusters, which could indicate that they are less engaged or have lower spending habits. Cluster 1 has higher values for all features, suggesting that these customers are more engaged and have higher spending habits. Cluster 2 has moderate values for all features, indicating that these customers may be somewhat engaged but not as much as those in Cluster 1. This information can be used to tailor marketing strategies and improve customer retention efforts.
```

```python
#Task 4
from sklearn.decomposition import PCA

# Apply PCA (2 components)
pca = PCA(n_components=2)
pca_data = pca.fit_transform(df_scaled)

# Convert to DataFrame
df_pca = pd.DataFrame(pca_data, columns=['PC1', 'PC2'])

# Explained variance ratio
print("Explained Variance Ratio:")
print(pca.explained_variance_ratio_)

# Feature loadings (components)
loadings = pd.DataFrame(
    pca.components_,
    columns=df.columns[:-1],  # exclude cluster column if added
    index=['PC1', 'PC2']
)

print("\nFeature Loadings:")
print(loadings)
```

```python
#Task 4 justification:
# PC1 and PC2 represent the two principal components that capture the most variance in the data. The explained variance ratio indicates how much of the total variance is captured by each principal component, which helps us understand the importance of each component. The feature loadings show how much each original feature contributes to the principal components, allowing us to interpret the underlying structure of the data and identify which features are most influential in differentiating between clusters.
#PC1 = Customer Value / Engagement Axis
#PC2 = Customer Activity / Recency Axis
```

```python
#Task 5
import matplotlib.pyplot as plt

# Combine PCA data with cluster labels
df_pca['cluster'] = df['cluster']

# Create scatter plot
plt.figure(figsize=(8,6))

for cluster in df_pca['cluster'].unique():
    subset = df_pca[df_pca['cluster'] == cluster]
    plt.scatter(subset['PC1'], subset['PC2'], label=f'Cluster {cluster}')

# Labels and title
plt.xlabel('Principal Component 1 (PC1)')
plt.ylabel('Principal Component 2 (PC2)')
plt.title('Customer Segmentation using K-Means (PCA Visualization)')

# Legend
plt.legend()

# Show plot
plt.show()
```
